4. Perform Hypothesis Testing to Make Statistical Inferences
• Formulate hypotheses and perform hypothesis testing to determine if body 
fat percentages differ significantly between genders.
• Calculate t-statistics and p-values to make informed decisions based on 
hypothesis test results.
• Evaluate the practical significance of the observed differences in body fat 
percentages

```python
#Formulating the Hypotheses

#Null Hypothesis (H0): There is no significant difference in body fat percentages between genders.
#Alternative Hypothesis (H1): There is a significant difference in body fat percentages between genders.
#Mathematically:

#H0: μ_male = μ_female (The mean body fat percentage of males is equal to the mean body fat percentage of females)
#H1: μ_male ≠ μ_female (The mean body fat percentage of males is not equal to the mean body fat percentage of females)


#Grouping the data into two groups based on gender and body fat%
fat_percent_male = body_performance_data[body_performance_data['gender'] == 'M']['body fat_%']
fat_percent_female = body_performance_data[body_performance_data['gender'] == 'F']['body fat_%']

#Calculating the statistics for the body fat% attribute in the dataset after grouping into male and female
mean_male_fat = np.mean(fat_percent_male)
mean_female_fat = np.mean(fat_percent_female)
std_male_fat = np.std(fat_percent_male)
std_female_fat = np.std(fat_percent_female)
alpha = 0.05

# Calculating  the pooled standard deviation for the body fat%
n_male = len(fat_percent_male)
n_female = len(fat_percent_female)
pooled_std_dev = np.sqrt(((n_male - 1) * std_male_fat**2 + (n_female - 1) * std_female_fat**2) / (n_male + n_female - 2))

# Calculating the t-statistic
t_statistic = (mean_male_fat - mean_female_fat) / (pooled_std_dev * np.sqrt(1/n_male + 1/n_female))
 # Calculating  the degrees of freedom
degrees_of_freedom = n_male + n_female - 2
# Calculating the p-value
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=degrees_of_freedom))
print("The t_statistic value  is :",t_statistic)
print("The p-value is            :",p_value)
# Interpreting the results
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in body fat percentages between genders.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in body fat percentages between genders.")
```

5. Estimate Confidence Intervals and Assess Sampling Variability in Data
• Estimate confidence intervals for the mean weight of individuals and 
interpret their significance.
• Assess how varying sample sizes affect the width of confidence intervals.
• Understand the impact of confidence levels on the precision of parameter 
estimates.


```python
#Calculating the statistics of the weight_kg attribute in the dtaset
mean_weight= np.mean(body_performance_data['weight_kg'])
std_weight = np.std(body_performance_data['weight_kg'])
# Setting confidence level 
confidence_level = 0.95
total_count=len(body_performance_data)
# Calculating critical value based on t-distribution 
alpha = 1 - confidence_level
critical_value = stats.t.ppf(1 - alpha / 2, df=total_count - 1)

# Calculating the standard error
standard_error = std_weight / np.sqrt(total_count)

# Calculating the confidence interval
lower_bound = mean_weight - critical_value * standard_error
upper_bound = mean_weight + critical_value * standard_error

# Interpreting the confidence interval
print(f"The {int(confidence_level * 100)}% Confidence Interval for Mean Weight: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python

# Defining a range of sample sizes
sample_sizes = [ 50, 500, 5000]
confidence_intervals = []
#Running  through a loop in sample_sizes
for count in sample_sizes:
    # Randomly selecting a sample of the specified size from your dataset
    sample_data = np.random.choice(body_performance_data['weight_kg'], size=count, replace=False)
    
    # Calculating sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # ddof=1 for sample standard deviation
    
    # Setting the confidence level (e.g., 95%)
    confidence_level = 0.95
    
    # Calculating the critical value based on t-distribution 
    alpha = 1 - confidence_level
    critical_value = stats.t.ppf(1 - alpha / 2, df=count - 1)
    
    # Calculating the standard error
    standard_error = sample_std / np.sqrt(count)
    
    # Calculating the  confidence interval
    lower_bound = sample_mean - critical_value * standard_error
    upper_bound = sample_mean + critical_value * standard_error
    
    confidence_intervals.append((count, lower_bound, upper_bound))
# Printing the confidence intervals for each sample size
for interval in confidence_intervals:
    sample_size, lower_bound, upper_bound = interval
    print(f"Sample Size: {sample_size}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})\n")
```

```python
 # Assigning Different confidence levels to explore 
confidence_levels = [0.90, 0.95, 0.99] 

# Creating lists to store confidence intervals
confidence_intervals = []
#looping through the given confidence levels and calculating the corresponding condidence intervals
for confidence_level in confidence_levels:
    alpha = 1 - confidence_level
    critical_value = stats.t.ppf(1 - alpha / 2, df=total_count - 1)
    standard_error = std_weight / np.sqrt(total_count)
    lower_bound = mean_weight - critical_value * standard_error
    upper_bound = mean_weight + critical_value * standard_error
    confidence_intervals.append((lower_bound , upper_bound))
# Printing the confidence intervals with their corresponding confidence levels
for i, confidence_level in enumerate(confidence_levels):
    print(f"Confidence Level: {confidence_level:.2f}")
    print(f"Confidence Interval: ({confidence_intervals[i][0]:.2f}, {confidence_intervals[i][1]:.2f})\n")
```
